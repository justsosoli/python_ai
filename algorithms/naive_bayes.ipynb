```python
from sklearn.datasets import fetch_20newsgroups

# data_home参数指定数据集下载和存储的本地目录
# subset参数指定加载数据的子集，'all'表示加载全部数据（既包含训练集也包含测试集）
newsgroups = fetch_20newsgroups(data_home="data", subset='all')

# 输出特征和标签
print("特征（文本数据）示例：")
print(newsgroups.data[0])
print("\n标签示例：", newsgroups.target[0])
print("标签总数：", len(newsgroups.target))
print("类别名称：", newsgroups.target_names)
print("data类型：", type(newsgroups.data))
print("target类型：", type(newsgroups.target))
print("target最小值：", min(newsgroups.target))
print("target最大值：", max(newsgroups.target))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# 将数据集分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(
    newsgroups.data, 
    newsgroups.target, 
    test_size=0.25, 
    random_state=42
)

# 使用TF-IDF向量化文本数据
# 这行代码的作用是创建一个TF-IDF向量化器（TfidfVectorizer），用于将文本数据转换为TF-IDF特征向量。
# 参数说明：
# stop_words='english'：移除英文停用词（如the、is等常见但没有实际意义的词）。
# max_df=0.5：忽略在超过50%文本中出现的高频词（这些词区分度低）。
# max_features=10000：只保留出现频率最高的10000个特征词，限制向量空间的维度。
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.5, max_features=10000)

X_train_tfidf = vectorizer.fit_transform(X_train)

print("处理后的特征数：", X_train_tfidf.shape[1])
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

X_test_tfidf = vectorizer.transform(X_test)

# 创建一个多项式朴素贝叶斯分类器对象，并将平滑参数alpha设置为0.01
nb = MultinomialNB(alpha=0.01)
nb.fit(X_train_tfidf, y_train)

y_pred = nb.predict(X_test_tfidf)

print("模型准确率：", accuracy_score(y_test, y_pred))
print("\n详细分类报告：\n", classification_report(y_test, y_pred, target_names=newsgroups.target_names))
```
